```python
def removeDuplicates(nums):
    if not nums:
        return []

    unique_index = 0

    for i in range(1, len(nums)):
        if nums[i] != nums[unique_index]:
            unique_index += 1
            nums[unique_index] = nums[i]

    return nums[:unique_index + 1]
```

```python
nums = [1,1,2,2,3,4,4]
print(removeDuplicates(nums))
```
