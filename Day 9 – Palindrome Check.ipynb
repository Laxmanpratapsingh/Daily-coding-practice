```python
def isPalindrome(s: str) -> bool:
    left, right = 0, len(s) - 1

    while left < right:
        if s[left] != s[right]:
            return False
        left += 1
        right -= 1

    return True
```

```python
print(isPalindrome("madam"))     # True
print(isPalindrome("racecar"))  # True
print(isPalindrome("hello"))    # False
print(isPalindrome("a"))        # True
print(isPalindrome(""))         # True
```
